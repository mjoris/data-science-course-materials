### Functies

```python
def ask_ok(prompt, retries=4, reminder='Please try again!'): # default args
    """Asks the user for confirmation""" # dit is de docstring
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)

ask_ok('OK to overwrite the file?', 2)
```

```python
ask_ok("is 't okee?", retries=0) # keyword arguments (steeds na postional args)
```

### \*args en \*\*kwargs

* __\*args__ verwijst naar een lijst van (onbenoemde) argumenten van een functie
* __\*\*kwargs__ verwijst naar een dictionary van keyword-argumenten van een functie
* kunnen als argument gebruikt worden van een functie

```python
def greetMe(name, **kwargs):  
    salutation = kwargs['sal'] if ('sal' in kwargs) else "Hallo" # omg it's a ternary operator
    print(salutation + ' ' + name)
  
greetMe('Joris', sal='Hoi')
```

* in principe kan om het even welke variabelenaam gebruikt worden, als de sterren er maar staan
* \*args en \*\*kwargs worden steevast gebruikt in de documentatie van libraries

### Anonieme functies met _lambda_

```python
add = lambda x, y: x + y
add(5, 6)
```

```python
a = [(1, 2), (4, 1), (9, 10), (13, -3)]
a.sort(key=lambda x: x[1])
print(a)
```

```python
items = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, items)) # map returns an iterator
print(squared)
odd = list(filter(lambda x: x % 2 == 1, squared)) # filter returns an iterator
print(odd)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib with labels, a title, ticks, a grid and a legend
plt.plot(x, y, 'r.')
plt.xlabel('tijd (minuten)')
plt.ylabel('hersenactiviteit (pulsen per seconde)')
plt.title('Aandachtscurve')
plt.xticks(range(0,10))
plt.grid(color='b', linestyle='--', linewidth=.5)
plt.legend(['Meting op vr 14/2'])
plt.show()
```
